```python
import pandas as pd

# Specify the path to your CSV file
file_path = f"tweets-data/{filename}"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=",")

# Display the DataFrame
display(df)
```

# Pre-Processing 1 : Membuang kolom yang tidak dibutuhkan

```python
#drop kolom yang tidak dibutuhkan (index kolom dimulai dari nol)
df.drop(df.columns[[0,1,2,4,5,6,7,8,9,10,11,13]], axis=1, inplace=True)

display(df)
```

# Pre-Processing2 : Membuat kolom"full_text"Menjadi LowerCase

```python
def casefoldingText(text): # Converting all the characters in a text into lower case
    text = text.lower()
    return text

df['caseFolded_fullText'] = df['full_text'].apply(casefoldingText)
df.head(200)
```

# Pre-Processing 3: Membersihkan text yang tidak diperlukan dengan menggunakan Reguler Expression (Regex)

```python
# Cleaning
import re
import string

def cleaningText(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text) # remove mentions
    text = re.sub(r'#[A-Za-z0-9]+', '', text) # remove hashtag
    text = re.sub(r'RT[\s]', '', text) # remove RT
    text = re.sub(r"http\S+", '', text) # remove link
    #text = re.sub(r'[0-9]+', '', text) # remove numbers
    text = re.sub(r'[^A-Za-z]+', '', text) # remove all character on alphabet

    text = text.replace('\n', ' ') # replace new line into space
    text = text.translate(str.maketrans('', '', string.punctuation)) # remove all punctuations
    text = text.strip(' ') # remove characters space from both left and right text
    return text

df['cleaning'] = df['caseFolded_fullText'].apply(cleaningText)
df.head(200)
```

# Pre-Procesing 4: Tokenisasi data untuk memisahkan kalimat menjadi kata per kata

```python
import nltk
nltk.download('punkt')
```

```python
# tokenisasi
from nltk.tokenize import word_tokenize

def tokenizingText(text): # Tokenizing or splitting a string, text into a list of token
    text = word_tokenize(text)
    return text

df['tokenizing'] = df['cleaning'].apply(tokenizingText)
# tweets = tweets[['text_clean', 'text_preprocessed']]
df.head(10)
```

# Pre-Processing 5: Mencari kata yang paling sering muncul

```python
from nltk import FreqDist

# Gabugkan semua kalimat dalam kolom 'Cleaning' menjadi satu string
kalimat = ' '.join(df['cleaning'])

tokens=nltk.tokenize.word_tokenize(kalimat)
kemunculan = nltk.FreqDist(tokens)
kemunculan.most_common(20)
```
